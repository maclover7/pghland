```python
import pandas as pd
import datetime
```

```python
# $('table.general-table > tbody > tr > td:first-child').map((idx, el) => el.innerHTML).toArray()

OFFENSES = [
    '601.03', '601.04', '601.06', '601.07', '601.08', '601.19', '604', '607.03', '607.05', '609', '613.02', '613.03', '613.04', '615', '616', '619.04', '633.08', '633.09', '633.11', '633.12', '633.20', '635.02',
    '4304', '5501', '5502', '5503', '5505', '5506', '5511', '5512', '5513', '5514', '5901', '5902', '5903', '5904', '6101', '6110.1', '6301', '6308', '6310', '6310.1', '6310.7', '6319', '6501'
]
```

```python
blotter = pd.concat([
    pd.read_csv('blotter-22.csv'),
    pd.read_csv('blotter-23.csv')
])
```

```python
blotter['INCIDENTTIME'] = pd.to_datetime(blotter.INCIDENTTIME)
blotter = blotter[blotter.INCIDENTTIME >= pd.to_datetime('today') - pd.Timedelta(days=365)]
```

```python
stdlocations = blotter[
    blotter.OFFENSES.str.contains('|'.join(OFFENSES))
][[
    'X', 'Y', 'INCIDENTLOCATION'
]].value_counts().reset_index().rename(columns={ 'INCIDENTLOCATION': 'STDINCIDENTLOCATION' })

stdlocations = stdlocations.drop_duplicates(subset=['X', 'Y'])
```

```python
blotter = pd.merge(
    blotter[blotter.OFFENSES.str.contains('|'.join(OFFENSES))],
    stdlocations[['X', 'Y', 'STDINCIDENTLOCATION']],
    on=['X', 'Y'],
    how='left')
```

```python
stdlocationsblottercount = blotter.STDINCIDENTLOCATION.value_counts().reset_index()
```

```python
stdlocationsblottercount[stdlocationsblottercount['count'] >= 3].to_csv('loc.csv')
```
